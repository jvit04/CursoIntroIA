```python
from sklearn import datasets

datos = datasets.load_breast_cancer(as_frame=True)
```

```python
from sklearn.model_selection import train_test_split
x=datos.data
y=datos.target
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.5)
```

```python
from math import gamma
from sklearn import svm
import sklearn.metrics as metrics

kernels = ['linear','rbf','sigmoid']
gamma = [1,.01,.001,.0001,.00001]
for kernel in kernels:
  for g in gamma:
    modelo = svm.SVC(kernel=kernel,gamma=g)
    modelo.fit(x_train,y_train)
    predicciones = modelo.predict(x_test)
    print(f"{kernel}: Gamma: {g}: Exactitud", metrics.accuracy_score(y_test,predicciones))
    print(f"{kernel}: Gamma: {g}: Precisión ", metrics.precision_score(y_test,predicciones))
```

```python
modelo = svm.SVC(kernel="linear")
modelo.fit(x_train,y_train)
predicciones = modelo.predict(x_test)
print(f"{kernel}: Gamma: {g}: Exactitud", metrics.accuracy_score(y_test,predicciones))
print(f"{kernel}: Gamma: {g}: Precisión ", metrics.precision_score(y_test,predicciones))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicciones))
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

pd.DataFrame(
    confusion_matrix(y_test,predicciones),
    columns=['Predice 0','Predice 1'],
    index=['Realmente 0','Realmente 1']
)
```
